```python
import libcst
```

```python
module_cst = libcst.parse_module("a = 10")
```

```python
module_cst
```

```python
from libcst import CSTTransformer
```

```python
class UpperNameTransformer(CSTTransformer):
    def leave_Name(self, origin, updated):
        return libcst.Name(value=origin.value.upper())
```

```python
modified_cst = module_cst.visit(UpperNameTransformer())
```

```python
modified_cst.code
```
