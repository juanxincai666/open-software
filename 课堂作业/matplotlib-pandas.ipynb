```python
x = np.arange(0, 10, 0.01)
```

```python
fig, ((ax00, ax01), (ax10, ax11)) = plt.subplots(2, 2)
ax00.plot(x, np.sin(x))
ax00.grid()
ax01.semilogx(x, np.sin(x))
ax01.grid()
ax10.semilogy(x, np.sin(x))
ax10.grid()
ax11.loglog(x, np.sin(x))
ax11.grid()
```

```python
with plt.xkcd():
    plt.plot(x, np.sin(x))
    plt.xlabel("x")
    plt.ylabel("sin(x)")
    plt.grid(True)
    plt.title("This is an xkcd-styled figure")
```
